# Hexbin plots

```python
%matplotlib inline

import pandas as pd
import numpy as np
import seaborn as sns

sns.set_theme(style="ticks")
```

I am using my favourite dataset, the IMDB movies dataset from kaggle.

```python
movies_df = pd.read_csv('C:/Users/ishwi/OneDrive/Documents/Python 50 days/Datasets/IMDb movies.csv',encoding = "ISO-8859-1")
```

Let's start with a scatter plot.

Although you can see that there are quite a few outliers for movie duration, it is clear that most movies are around the 200 mins mark in terms of length. I would analyse this data further using box plot and statistical summary of the variable but for this tutorial I want to focus on scatter plot.

```python
movies_df.plot.scatter(x='avg_vote', y='duration', title = 'Avg_vote vs Duration for movies')
```

The hexbin plot reveals that most movies are between 100 to 200 mins and have avg_votes between 4 to 8. This insight was hidden by the scatter plot.

```python
movies_df.plot.hexbin(x='avg_vote', y='duration', gridsize=20)
```

```python
movies_df.plot.hexbin(x='avg_vote', y='duration', gridsize=20, bins= "log")
```

My next step in this analysis would be to cut the data into bins and create cross tabs to analyse the actual numbers. But the following plot in seaborn may just be helpful before I jump into that.

```python
sns.jointplot(x=movies_df['avg_vote'], y=movies_df['duration'], kind="hex", color="#4CB391")
```

For now, I just wanted to demonstrate the use case of a hexbin plot and its advantage over a scatter plot.
